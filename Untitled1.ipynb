```python
import pickle

for fname in ["data/09_SY.pickle", "data/16_GZ.pickle", "data/17_JP.pickle"]:
    print("\n====== Checking:", fname, "======")
    with open(fname, "rb") as f:
        obj = pickle.load(f)
    print("Type:", type(obj))

    if isinstance(obj, dict):
        print("Keys:", obj.keys())

        for k, v in obj.items():
            if hasattr(v, "shape"):
                print("  ", k, "shape:", v.shape)
            else:
                print("  ", k, "type:", type(v))

    elif isinstance(obj, list):
        print("List length:", len(obj))
        print("First item type:", type(obj[0]))
        
        if isinstance(obj[0], dict):
            print("Item keys:", obj[0].keys())
            for k, v in obj[0].items():
                if hasattr(v, "shape"):
                    print("  ", k, "shape:", v.shape)
                else:
                    print("  ", k, "type:", type(v))

    else:
        print("Unknown structure, example value:", obj)
```
